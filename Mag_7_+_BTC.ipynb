```python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„¤ì¹˜ ë° ì„í¬íŠ¸
!pip install yfinance pandas numpy plotly -q

import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio
warnings.filterwarnings('ignore')

# êµ¬ê¸€ ì½”ë©ìš© ë Œë”ëŸ¬ ì„¤ì •
pio.renderers.default = 'colab'

# ==================== Magnificent Seven + BTC ì •ì˜ ====================

MAG7_STOCKS = {
    'AAPL': {
        'name': 'Apple Inc.',
        'description': 'ì•„ì´í°, ìƒíƒœê³„, ì˜¨ë””ë°”ì´ìŠ¤ AI',
        'sector': 'Technology',
        'industry': 'Consumer Electronics',
        'type': 'Stock'
    },
    'MSFT': {
        'name': 'Microsoft Corporation',
        'description': 'í´ë¼ìš°ë“œ(Azure), ìƒì„±í˜• AI (OpenAI ëŒ€ì£¼ì£¼)',
        'sector': 'Technology',
        'industry': 'Software',
        'type': 'Stock'
    },
    'GOOGL': {
        'name': 'Alphabet Inc.',
        'description': 'êµ¬ê¸€ ê²€ìƒ‰, ìœ íŠœë¸Œ, AI (Gemini)',
        'sector': 'Communication Services',
        'industry': 'Internet Content & Information',
        'type': 'Stock'
    },
    'AMZN': {
        'name': 'Amazon.com Inc.',
        'description': 'ì „ììƒê±°ë˜, í´ë¼ìš°ë“œ(AWS) 1ìœ„',
        'sector': 'Consumer Cyclical',
        'industry': 'Internet Retail',
        'type': 'Stock'
    },
    'NVDA': {
        'name': 'NVIDIA Corporation',
        'description': 'AI ë°˜ë„ì²´(GPU) ë…ì ì  ì§€ë°°ì',
        'sector': 'Technology',
        'industry': 'Semiconductors',
        'type': 'Stock'
    },
    'META': {
        'name': 'Meta Platforms Inc.',
        'description': 'í˜ì´ìŠ¤ë¶, ì¸ìŠ¤íƒ€ê·¸ë¨, AI(Llama)',
        'sector': 'Communication Services',
        'industry': 'Internet Content & Information',
        'type': 'Stock'
    },
    'TSLA': {
        'name': 'Tesla Inc.',
        'description': 'ì „ê¸°ì°¨, ììœ¨ì£¼í–‰, ë¡œë´‡',
        'sector': 'Consumer Cyclical',
        'industry': 'Auto Manufacturers',
        'type': 'Stock'
    },
    'BTC-USD': {
        'name': 'Bitcoin',
        'description': 'ë””ì§€í„¸ ê³¨ë“œ, íƒˆì¤‘ì•™í™” í™”í, ê°€ì¹˜ ì €ì¥ ìˆ˜ë‹¨',
        'sector': 'Cryptocurrency',
        'industry': 'Digital Assets',
        'type': 'Crypto'
    }
}

print("=" * 80)
print("ğŸŒŸ MAGNIFICENT SEVEN + BITCOIN ì‹¤ì‹œê°„ ì¢…í•© ë¶„ì„")
print("=" * 80)
print("\në¹…í…Œí¬ 7ëŒ€ ê¸°ì—… + ë¹„íŠ¸ì½”ì¸:")
for ticker, info in MAG7_STOCKS.items():
    icon = "â‚¿" if info['type'] == 'Crypto' else "ğŸ“ˆ"
    print(f"  {icon} {ticker:8} - {info['name']:30} | {info['description']}")
print("=" * 80)

def get_current_quarter_start():
    """í˜„ì¬ ë¶„ê¸° ì‹œì‘ì¼ ê³„ì‚°"""
    now = datetime.now()
    quarter = (now.month - 1) // 3
    quarter_start_month = quarter * 3 + 1
    quarter_start = datetime(now.year, quarter_start_month, 1)
    return quarter_start

def calculate_anchored_vwap(df):
    """Anchored VWAP ê³„ì‚°"""
    df = df.copy()
    df['Typical_Price'] = (df['High'] + df['Low'] + df['Close']) / 3
    df['TP_Volume'] = df['Typical_Price'] * df['Volume']
    df['Cumulative_TP_Volume'] = df['TP_Volume'].cumsum()
    df['Cumulative_Volume'] = df['Volume'].cumsum()
    df['Anchored_VWAP'] = df['Cumulative_TP_Volume'] / df['Cumulative_Volume']
    return df

def get_quarterly_vwap_analysis(ticker):
    """ë¶„ê¸°ë³„ Anchored VWAP ë¶„ì„"""
    try:
        quarter_start = get_current_quarter_start()
        end_date = datetime.now()
        quarter_num = (quarter_start.month - 1) // 3 + 1

        stock = yf.Ticker(ticker)
        df = stock.history(start=quarter_start, end=end_date)

        if df.empty or len(df) < 5:
            return None

        df = calculate_anchored_vwap(df)

        current_price = df['Close'].iloc[-1]
        current_vwap = df['Anchored_VWAP'].iloc[-1]
        above_vwap_ratio = (df['Close'] > df['Anchored_VWAP']).sum() / len(df) * 100
        recent_5days_avg = df['Close'].tail(5).mean()
        recent_10days_avg = df['Close'].tail(10).mean()

        recent_20 = df['Close'].tail(min(20, len(df)))
        uptrend_strength = (recent_20.diff() > 0).sum() / len(recent_20) * 100 if len(recent_20) > 1 else 50

        recent_volume = df['Volume'].tail(5).mean()
        avg_volume = df['Volume'].mean()
        volume_ratio = recent_volume / avg_volume if avg_volume > 0 else 1

        info = stock.info
        quarter_start_price = df['Close'].iloc[0]
        quarter_return = ((current_price - quarter_start_price) / quarter_start_price * 100)

        # ì‹œê°€ì´ì•¡ ì²˜ë¦¬ (ë¹„íŠ¸ì½”ì¸ì˜ ê²½ìš° íŠ¹ìˆ˜ ì²˜ë¦¬)
        if ticker == 'BTC-USD':
            # ë¹„íŠ¸ì½”ì¸ ì‹œê°€ì´ì•¡ = í˜„ì¬ê°€ Ã— ìœ í†µëŸ‰
            # Yahoo Financeì—ì„œ ì œê³µí•˜ëŠ” ê²½ìš°ê°€ ìˆì§€ë§Œ, ì—†ìœ¼ë©´ ëŒ€ëµ ê³„ì‚°
            market_cap = info.get('marketCap', current_price * 19.5e6 * 1e9)  # ì•½ 1950ë§Œ BTC ìœ í†µ
        else:
            market_cap = info.get('marketCap', 0)

        return {
            'Ticker': ticker,
            'Company': MAG7_STOCKS[ticker]['name'],
            'Description': MAG7_STOCKS[ticker]['description'],
            'Sector': MAG7_STOCKS[ticker]['sector'],
            'Type': MAG7_STOCKS[ticker]['type'],
            'Quarter': f'{quarter_start.year} Q{quarter_num}',
            'Quarter_Start_Date': quarter_start.strftime('%Y-%m-%d'),
            'Trading_Days': len(df),
            'Current_Price': round(current_price, 2),
            'Anchored_VWAP': round(current_vwap, 2),
            'Quarter_Start_Price': round(quarter_start_price, 2),
            'Quarter_Return_%': round(quarter_return, 2),
            'Price_vs_VWAP_%': round((current_price - current_vwap) / current_vwap * 100, 2),
            'Above_VWAP_Days_%': round(above_vwap_ratio, 1),
            'Recent_5D_Avg': round(recent_5days_avg, 2),
            'Recent_10D_Avg': round(recent_10days_avg, 2),
            'Uptrend_Strength_%': round(uptrend_strength, 1),
            'Volume_Ratio': round(volume_ratio, 2),
            'Is_Above_VWAP': current_price > current_vwap,
            'Market_Cap': market_cap,
            'Buy_Signal_Score': 0
        }

    except Exception as e:
        print(f"Error processing {ticker}: {str(e)}")
        return None

# ë¶„ê¸° ì •ë³´
quarter_start = get_current_quarter_start()
quarter_num = (quarter_start.month - 1) // 3 + 1

print("\n" + "=" * 80)
print(f"ğŸ¯ Anchored VWAP ë¶„ì„: {quarter_start.year} Q{quarter_num}")
print(f"ğŸ“ Anchor Point: {quarter_start.strftime('%Y-%m-%d')}")
print(f"ğŸ“… í˜„ì¬ê¹Œì§€ ê±°ë˜ì¼: {(datetime.now() - quarter_start).days}ì¼")
print("=" * 80)

# MAG 7 + BTC ë¶„ì„
results = []
all_tickers = list(MAG7_STOCKS.keys())

print("\nğŸ” MAG 7 + BTC ì¢…ëª© ë¶„ì„ ì¤‘...\n")

for ticker in all_tickers:
    icon = "â‚¿" if MAG7_STOCKS[ticker]['type'] == 'Crypto' else "ğŸ“ˆ"
    print(f"ë¶„ì„ ì¤‘: {icon} {ticker} ({MAG7_STOCKS[ticker]['name']})...", end=" ")
    result = get_quarterly_vwap_analysis(ticker)
    if result:
        results.append(result)
        print("âœ“")
    else:
        print("âœ—")

df_results = pd.DataFrame(results)

# ë§¤ìˆ˜ ì‹ í˜¸ ì ìˆ˜ ê³„ì‚°
def calculate_buy_score(row):
    score = 0
    if row['Is_Above_VWAP']:
        score += 30

    price_diff = row['Price_vs_VWAP_%']
    if 0 < price_diff <= 5:
        score += 20
    elif 5 < price_diff <= 10:
        score += 10
    elif price_diff > 10:
        score += 5

    if row['Above_VWAP_Days_%'] >= 80:
        score += 20
    elif row['Above_VWAP_Days_%'] >= 60:
        score += 15
    elif row['Above_VWAP_Days_%'] >= 40:
        score += 10

    if row['Uptrend_Strength_%'] >= 60:
        score += 15
    elif row['Uptrend_Strength_%'] >= 50:
        score += 10

    if row['Volume_Ratio'] >= 1.2:
        score += 15
    elif row['Volume_Ratio'] >= 1.0:
        score += 10

    return min(score, 100)

df_results['Buy_Signal_Score'] = df_results.apply(calculate_buy_score, axis=1)
df_results = df_results.sort_values('Buy_Signal_Score', ascending=False)

# ì‹œê°€ì´ì•¡ ì¶”ê°€
df_results['Market_Cap_Trillion'] = (df_results['Market_Cap'] / 1e12).round(3)

print("\n" + "=" * 80)
print("ğŸ“Š MAGNIFICENT SEVEN + BITCOIN - ì‹¤ì‹œê°„ ë¶„ì„ ê²°ê³¼")
print("=" * 80)

display_cols = ['Ticker', 'Company', 'Type', 'Description', 'Current_Price', 'Anchored_VWAP',
                'Price_vs_VWAP_%', 'Quarter_Return_%', 'Market_Cap_Trillion', 'Buy_Signal_Score']
print(df_results[display_cols].to_string(index=False))

print("\n" + "=" * 80)
print("ğŸ† MAG 7 + BTC íˆ¬ì ì¶”ì²œ ìˆœìœ„")
print("=" * 80)

for idx, row in df_results.iterrows():
    rank = df_results.index.get_loc(idx) + 1

    if rank == 1:
        medal = "ğŸ¥‡"
    elif rank == 2:
        medal = "ğŸ¥ˆ"
    elif rank == 3:
        medal = "ğŸ¥‰"
    else:
        medal = f"  {rank}"

    icon = "â‚¿" if row['Type'] == 'Crypto' else "ğŸ“ˆ"

    print(f"\n{medal} {icon} {row['Ticker']} - {row['Company']}")
    print(f"    ğŸ¯ {row['Description']}")
    print(f"    ğŸ’° ì‹œê°€ì´ì•¡: ${row['Market_Cap_Trillion']:.2f}T")

    if row['Type'] == 'Crypto':
        print(f"    ğŸ“ˆ í˜„ì¬ê°€: ${row['Current_Price']:,.0f} | VWAP: ${row['Anchored_VWAP']:,.0f}")
    else:
        print(f"    ğŸ“ˆ í˜„ì¬ê°€: ${row['Current_Price']} | VWAP: ${row['Anchored_VWAP']}")

    print(f"    ğŸ“Š VWAP ëŒ€ë¹„: {row['Price_vs_VWAP_%']:+.2f}% | ë¶„ê¸° ìˆ˜ìµë¥ : {row['Quarter_Return_%']:+.2f}%")
    print(f"    ğŸ² VWAP ìœ„ ê±°ë˜ì¼: {row['Above_VWAP_Days_%']}% | ì¶”ì„¸ ê°•ë„: {row['Uptrend_Strength_%']}%")
    print(f"    â­ ë§¤ìˆ˜ ì‹ í˜¸ ì ìˆ˜: {row['Buy_Signal_Score']}/100", end="")

    if row['Buy_Signal_Score'] >= 80:
        print(" â†’ ğŸ’š ê°•ë ¥ ë§¤ìˆ˜ ì¶”ì²œ")
    elif row['Buy_Signal_Score'] >= 60:
        print(" â†’ ğŸ’› ëˆŒë¦¼ëª© ëŒ€ê¸°")
    else:
        print(" â†’ ğŸ’™ ê´€ì°° í•„ìš”")

# ==================== í€ë”ë©˜í„¸ ë¶„ì„ ====================

def get_comprehensive_analysis(ticker):
    """ì¢…ëª©ë³„ í€ë”ë©˜í„¸ ë¶„ì„"""
    try:
        stock = yf.Ticker(ticker)
        info = stock.info
        current_price = info.get('currentPrice', info.get('regularMarketPrice', 0))

        # ë¹„íŠ¸ì½”ì¸ì¸ ê²½ìš° íŠ¹ìˆ˜ ì²˜ë¦¬
        if ticker == 'BTC-USD':
            report = {
                "===== [ê¸°ë³¸ ì •ë³´] =====": "",
                "ìì‚°ëª…": "Bitcoin (BTC)",
                "ìœ í˜•": "ì•”í˜¸í™”í (Cryptocurrency)",
                "ì„¤ëª…": "íƒˆì¤‘ì•™í™” ë””ì§€í„¸ í™”í, ë””ì§€í„¸ ê³¨ë“œ",
                "í˜„ì¬ê°€": f"${current_price:,.2f}",
                "ì‹œê°€ì´ì•¡": f"${info.get('marketCap', 0) / 1e12:.3f}T" if info.get('marketCap') else "N/A",

                "===== [íŠ¹ì„±] =====": "",
                "ì´ ë°œí–‰ëŸ‰": "21,000,000 BTC (ê³ ì •)",
                "í˜„ì¬ ìœ í†µëŸ‰": "ì•½ 19,500,000 BTC",
                "ë°˜ê°ê¸°": "4ë…„ë§ˆë‹¤ (ë‹¤ìŒ: 2028ë…„)",
                "ì»¨ì„¼ì„œìŠ¤": "Proof of Work (PoW)",

                "===== [íˆ¬ì í¬ì¸íŠ¸] =====": "",
                "ì¥ì ": "ì¸í”Œë ˆì´ì…˜ í—¤ì§€, ê¸°ê´€ ìˆ˜ìš” ì¦ê°€, ETF ìŠ¹ì¸",
                "ë¦¬ìŠ¤í¬": "ë³€ë™ì„±, ê·œì œ ë¶ˆí™•ì‹¤ì„±, ì—ë„ˆì§€ ì†Œë¹„",
                "í¬ì§€ì…”ë‹": "ëŒ€ì²´ ìì‚°, í¬íŠ¸í´ë¦¬ì˜¤ ë‹¤ê°í™”",
            }
            return report

        def safe_get(key, default='N/A', multiplier=1, format_str=None):
            value = info.get(key)
            if value is None or value == 'N/A':
                return default
            try:
                if format_str:
                    return format_str.format(value * multiplier)
                return value * multiplier
            except:
                return default

        recommendation_map = {
            'buy': 'ë§¤ìˆ˜', 'strong buy': 'ì ê·¹ ë§¤ìˆ˜', 'hold': 'ë³´ìœ ',
            'sell': 'ë§¤ë„', 'strong sell': 'ì ê·¹ ë§¤ë„'
        }
        rec_key = info.get('recommendationKey', 'N/A').lower()
        recommendation_kr = recommendation_map.get(rec_key, rec_key.upper())

        target_price = safe_get('targetMeanPrice', 0)
        upside = 'N/A'
        if target_price and target_price > 0 and current_price > 0:
            upside = f"{((target_price / current_price) - 1) * 100:.2f}%"

        report = {
            "===== [ê¸°ë³¸ ì •ë³´] =====": "",
            "ì¢…ëª©ëª…": info.get('longName', ticker),
            "ë¹„ì¦ˆë‹ˆìŠ¤": MAG7_STOCKS[ticker]['description'],
            "ì„¹í„°": info.get('sector', 'N/A'),
            "í˜„ì¬ê°€": f"${current_price:.2f}" if current_price else "N/A",
            "ì‹œê°€ì´ì•¡": f"${safe_get('marketCap', 0) / 1e12:.3f}T" if safe_get('marketCap', 0) else "N/A",

            "===== [ê°€ì¹˜ í‰ê°€] =====": "",
            "PER (ì£¼ê°€ìˆ˜ìµë¹„ìœ¨)": f"{safe_get('trailingPE', 0):.2f}" if safe_get('trailingPE') != 'N/A' else "N/A",
            "Forward PER": f"{safe_get('forwardPE', 0):.2f}" if safe_get('forwardPE') != 'N/A' else "N/A",
            "PBR": f"{safe_get('priceToBook', 0):.2f}" if safe_get('priceToBook') != 'N/A' else "N/A",
            "PEG Ratio": f"{safe_get('pegRatio', 0):.2f}" if safe_get('pegRatio') != 'N/A' else "N/A",

            "===== [ìˆ˜ìµì„± & ì¬ë¬´] =====": "",
            "ROE": safe_get('returnOnEquity', default='N/A', multiplier=100, format_str="{:.2f}%"),
            "ì˜ì—…ì´ìµë¥ ": safe_get('operatingMargins', default='N/A', multiplier=100, format_str="{:.2f}%"),
            "ìˆœì´ìµë¥ ": safe_get('profitMargins', default='N/A', multiplier=100, format_str="{:.2f}%"),
            "ë¶€ì±„ë¹„ìœ¨": f"{safe_get('debtToEquity', 0):.2f}%" if safe_get('debtToEquity') != 'N/A' else "N/A",

            "===== [ì„±ì¥ì„±] =====": "",
            "ë§¤ì¶œ ì„±ì¥ë¥ ": safe_get('revenueGrowth', default='N/A', multiplier=100, format_str="{:.2f}%"),
            "ì´ìµ ì„±ì¥ë¥ ": safe_get('earningsGrowth', default='N/A', multiplier=100, format_str="{:.2f}%"),

            "===== [ì›”ê°€ ì»¨ì„¼ì„œìŠ¤] =====": "",
            "íˆ¬ìì˜ê²¬": recommendation_kr,
            "ëª©í‘œì£¼ê°€": f"${target_price:.2f}" if target_price else "N/A",
            "ìƒìŠ¹ì—¬ë ¥": upside,
        }

        return report

    except Exception as e:
        return {"Error": f"ë¶„ì„ ì‹¤íŒ¨: {str(e)}"}

print("\n\n" + "=" * 80)
print("ğŸ“‹ MAG 7 + BTC - í€ë”ë©˜í„¸ ìƒì„¸ ë¶„ì„")
print("=" * 80)

for idx, row in df_results.iterrows():
    ticker = row['Ticker']
    rank = df_results.index.get_loc(idx) + 1
    icon = "â‚¿" if row['Type'] == 'Crypto' else "ğŸ“ˆ"

    print(f"\n{'='*80}")
    print(f"[ìˆœìœ„ {rank}] {icon} {ticker} - {row['Company']}")
    print("="*80)

    analysis = get_comprehensive_analysis(ticker)

    for key, value in analysis.items():
        if "=====" in key:
            print(f"\n{value}{key}")
        else:
            print(f"  {key:<20}: {value}")

    print("-" * 80)

# ==================== ì°¨íŠ¸ ìƒì„± ====================

print("\nğŸ“ˆ ì¸í„°ë™í‹°ë¸Œ ì°¨íŠ¸ ìƒì„± ì¤‘...")

# ì°¨íŠ¸ 1: ë§¤ìˆ˜ ì‹ í˜¸ ì ìˆ˜
fig1 = go.Figure()

colors_score = []
for idx, row in df_results.iterrows():
    if row['Buy_Signal_Score'] >= 80:
        color = '#00C853'
    elif row['Buy_Signal_Score'] >= 60:
        color = '#FFD600'
    else:
        color = '#2196F3'

    # ë¹„íŠ¸ì½”ì¸ì€ ì˜¤ë Œì§€ìƒ‰ìœ¼ë¡œ ê°•ì¡°
    if row['Type'] == 'Crypto':
        color = '#FF9800'

    colors_score.append(color)

fig1.add_trace(go.Bar(
    y=df_results['Ticker'],
    x=df_results['Buy_Signal_Score'],
    orientation='h',
    marker=dict(color=colors_score),
    text=df_results['Buy_Signal_Score'],
    textposition='auto',
    hovertemplate='<b>%{y}</b><br>ë§¤ìˆ˜ ì ìˆ˜: %{x}/100<extra></extra>'
))

fig1.update_layout(
    title=f'ğŸ¯ MAG 7 + BTC ë§¤ìˆ˜ ì‹ í˜¸ ì ìˆ˜ (Anchored VWAP - {quarter_start.year} Q{quarter_num})',
    xaxis_title='ë§¤ìˆ˜ ì‹ í˜¸ ì ìˆ˜',
    yaxis_title='ì¢…ëª©',
    height=550,
    template='plotly_white'
)

fig1.show()

# ì°¨íŠ¸ 2: VWAP ëŒ€ë¹„ ê°€ê²© ìœ„ì¹˜
fig2 = go.Figure()

colors = []
for idx, row in df_results.iterrows():
    if row['Price_vs_VWAP_%'] > 0:
        color = '#FF9800' if row['Type'] == 'Crypto' else 'green'
    else:
        color = '#FF6B6B' if row['Type'] == 'Crypto' else 'red'
    colors.append(color)

fig2.add_trace(go.Bar(
    y=df_results['Ticker'],
    x=df_results['Price_vs_VWAP_%'],
    orientation='h',
    marker=dict(color=colors),
    text=df_results['Price_vs_VWAP_%'].round(1),
    textposition='auto'
))

fig2.add_vline(x=0, line_dash="dash", line_color="black", line_width=2)

fig2.update_layout(
    title=f'ğŸ“Š MAG 7 + BTC - Anchored VWAP ëŒ€ë¹„ ê°€ê²© ìœ„ì¹˜ ({quarter_start.year} Q{quarter_num})',
    xaxis_title='VWAP ëŒ€ë¹„ ì°¨ì´ (%)',
    yaxis_title='ì¢…ëª©',
    height=550,
    template='plotly_white'
)

fig2.show()

# ì°¨íŠ¸ 3: ì‹œê°€ì´ì•¡ vs ë¶„ê¸° ìˆ˜ìµë¥ 
fig3 = px.scatter(
    df_results,
    x='Market_Cap_Trillion',
    y='Quarter_Return_%',
    size='Buy_Signal_Score',
    color='Type',
    hover_data=['Ticker', 'Company', 'Description'],
    text='Ticker',
    color_discrete_map={'Stock': '#2196F3', 'Crypto': '#FF9800'},
    title=f'ğŸ’° MAG 7 + BTC - ì‹œê°€ì´ì•¡ vs ë¶„ê¸° ìˆ˜ìµë¥  ({quarter_start.year} Q{quarter_num})',
    labels={'Market_Cap_Trillion': 'ì‹œê°€ì´ì•¡ (Trillion USD)',
            'Quarter_Return_%': 'ë¶„ê¸° ìˆ˜ìµë¥  (%)'}
)

fig3.update_traces(textposition='top center', textfont_size=12)
fig3.update_layout(height=600)
fig3.show()

# ì°¨íŠ¸ 4: ê° ì¢…ëª©ë³„ 1ë…„ ì°¨íŠ¸
print("\nğŸ“ˆ MAG 7 + BTC ì¢…ëª©ë³„ 1ë…„ ì°¨íŠ¸ ìƒì„± ì¤‘...")

def get_quarterly_anchors(start_date, end_date):
    """ë¶„ê¸° ì‹œì‘ì¼ ê³„ì‚°"""
    quarters = []
    current = start_date

    while current <= end_date:
        year = current.year
        month = current.month
        quarter_start_month = ((month - 1) // 3) * 3 + 1
        quarter_start = datetime(year, quarter_start_month, 1)

        if quarter_start not in [q[0] for q in quarters] and quarter_start >= start_date:
            quarter_num = (quarter_start_month - 1) // 3 + 1
            quarters.append((quarter_start, f"Q{quarter_num} {year}"))

        if month >= 10:
            current = datetime(year + 1, 1, 1)
        else:
            current = datetime(year, quarter_start_month + 3, 1)

    return quarters

def calculate_multiple_anchored_vwaps(df, anchor_points):
    """ì—¬ëŸ¬ ë¶„ê¸° VWAP ê³„ì‚°"""
    df = df.copy()
    df['Typical_Price'] = (df['High'] + df['Low'] + df['Close']) / 3
    vwap_columns = {}
    df_tz = df.index.tz if hasattr(df.index, 'tz') else None

    for anchor_date, quarter_label in anchor_points:
        try:
            if df_tz is not None:
                anchor_date_tz = pd.Timestamp(anchor_date).tz_localize(df_tz)
            else:
                anchor_date_tz = pd.Timestamp(anchor_date)

            mask = df.index >= anchor_date_tz
            if mask.sum() == 0:
                continue

            df_period = df[mask].copy()
            tp_volume = (df_period['Typical_Price'] * df_period['Volume']).cumsum()
            cumulative_volume = df_period['Volume'].cumsum()
            vwap = tp_volume / cumulative_volume

            vwap_full = pd.Series(index=df.index, dtype=float)
            vwap_full[mask] = vwap.values
            vwap_columns[quarter_label] = vwap_full

        except Exception as e:
            continue

    return vwap_columns

# ê° ì¢…ëª©ë³„ ì°¨íŠ¸
for idx, row in df_results.iterrows():
    ticker = row['Ticker']
    rank = df_results.index.get_loc(idx) + 1
    icon = "â‚¿" if row['Type'] == 'Crypto' else "ğŸ“ˆ"

    print(f"\n[{rank}/8] ì°¨íŠ¸ ìƒì„±: {icon} {ticker}...", end=" ")

    try:
        end_date = datetime.now()
        start_date = end_date - timedelta(days=365)

        stock = yf.Ticker(ticker)
        df_1year = stock.history(start=start_date, end=end_date)

        if df_1year.empty:
            print("âœ— ë°ì´í„° ì—†ìŒ")
            continue

        quarter_anchors = get_quarterly_anchors(start_date, end_date)
        vwap_dict = calculate_multiple_anchored_vwaps(df_1year, quarter_anchors)

        fig = go.Figure()

        # ìº”ë“¤ìŠ¤í‹±
        fig.add_trace(go.Candlestick(
            x=df_1year.index,
            open=df_1year['Open'],
            high=df_1year['High'],
            low=df_1year['Low'],
            close=df_1year['Close'],
            name='Price',
            increasing_line_color='#26a69a',
            decreasing_line_color='#ef5350'
        ))

        # ë¶„ê¸°ë³„ VWAP
        colors_vwap = ['#2196F3', '#FF9800', '#4CAF50', '#9C27B0', '#F44336']

        for idx_vwap, (quarter_label, vwap_series) in enumerate(vwap_dict.items()):
            color = colors_vwap[idx_vwap % len(colors_vwap)]
            vwap_clean = vwap_series.dropna()

            if len(vwap_clean) > 0:
                fig.add_trace(go.Scatter(
                    x=vwap_clean.index,
                    y=vwap_clean,
                    mode='lines',
                    name=f'VWAP {quarter_label}',
                    line=dict(color=color, width=2),
                    opacity=0.7
                ))

        # í˜„ì¬ ë¶„ê¸° ê°•ì¡°
        if vwap_dict:
            latest_quarter = list(vwap_dict.keys())[-1]
            latest_vwap = vwap_dict[latest_quarter].dropna()

            if len(latest_vwap) > 0:
                fig.add_trace(go.Scatter(
                    x=latest_vwap.index,
                    y=latest_vwap,
                    mode='lines',
                    name=f'âœ¨ Current {latest_quarter}',
                    line=dict(color='red', width=3),
                    opacity=1
                ))

        # ê±°ë˜ëŸ‰
        fig.add_trace(go.Bar(
            x=df_1year.index,
            y=df_1year['Volume'],
            name='Volume',
            marker_color='rgba(128, 128, 128, 0.3)',
            yaxis='y2'
        ))

        # ê°€ê²© í¬ë§· (ë¹„íŠ¸ì½”ì¸ì€ ì²œ ë‹¨ìœ„ êµ¬ë¶„)
        if row['Type'] == 'Crypto':
            price_format = f"${row['Current_Price']:,.0f}"
            vwap_format = f"${row['Anchored_VWAP']:,.0f}"
        else:
            price_format = f"${row['Current_Price']}"
            vwap_format = f"${row['Anchored_VWAP']}"

        fig.update_layout(
            title={
                'text': f"<b>#{rank} {icon} {ticker} - {row['Company']}</b><br>" +
                        f"<sub>{row['Description']}</sub><br>" +
                        f"<sub>ë§¤ìˆ˜ ì ìˆ˜: {row['Buy_Signal_Score']}/100 | " +
                        f"í˜„ì¬ê°€: {price_format} | " +
                        f"VWAP: {vwap_format} | " +
                        f"ì°¨ì´: {row['Price_vs_VWAP_%']:+.2f}%</sub>",
                'x': 0.5,
                'xanchor': 'center'
            },
            xaxis=dict(title='ë‚ ì§œ', rangeslider=dict(visible=False)),
            yaxis=dict(title='ê°€ê²© (USD)', side='right'),
            yaxis2=dict(
                title='ê±°ë˜ëŸ‰',
                overlaying='y',
                side='left',
                showgrid=False,
                range=[0, df_1year['Volume'].max() * 4]
            ),
            hovermode='x unified',
            height=700,
            template='plotly_white'
        )

        # ë¶„ê¸° ì„¸ë¡œì„ 
        df_tz = df_1year.index.tz if hasattr(df_1year.index, 'tz') else None

        for anchor_date, quarter_label in quarter_anchors:
            if df_tz is not None:
                anchor_display = pd.Timestamp(anchor_date).tz_localize(df_tz)
            else:
                anchor_display = anchor_date

            fig.add_shape(
                type="line",
                x0=anchor_display,
                x1=anchor_display,
                y0=0,
                y1=1,
                yref="paper",
                line=dict(color="gray", width=1, dash="dash"),
                opacity=0.5
            )

            fig.add_annotation(
                x=anchor_display,
                y=1,
                yref="paper",
                text=quarter_label,
                showarrow=False,
                yshift=10,
                font=dict(size=10, color="gray")
            )

        fig.show()
        print("âœ“")

    except Exception as e:
        print(f"âœ— ì‹¤íŒ¨: {str(e)}")

# ì°¨íŠ¸ 5: MAG 7 + BTC ë¹„êµ ì°¨íŠ¸
print("\nğŸ“Š MAG 7 + BTC ìˆ˜ìµë¥  ë¹„êµ ì°¨íŠ¸ ìƒì„± ì¤‘...")

fig_compare = go.Figure()

for ticker in all_tickers:
    try:
        end_date = datetime.now()
        start_date = end_date - timedelta(days=365)

        stock = yf.Ticker(ticker)
        df_1year = stock.history(start=start_date, end=end_date)

        if df_1year.empty:
            continue

        normalized_price = (df_1year['Close'] / df_1year['Close'].iloc[0]) * 100

        # ë¹„íŠ¸ì½”ì¸ì€ ì ì„ ìœ¼ë¡œ êµ¬ë¶„
        line_style = dict(width=3, dash='dash') if ticker == 'BTC-USD' else dict(width=2.5)

        fig_compare.add_trace(go.Scatter(
            x=df_1year.index,
            y=normalized_price,
            mode='lines',
            name=ticker,
            line=line_style,
            hovertemplate=f'<b>{ticker}</b><br>ë‚ ì§œ: %{{x}}<br>ìˆ˜ìµë¥ : %{{y:.2f}}%<extra></extra>'
        ))

    except:
        continue

fig_compare.update_layout(
    title='ğŸ† MAGNIFICENT SEVEN + BITCOIN - 1ë…„ ìˆ˜ìµë¥  ë¹„êµ (ì •ê·œí™”)',
    xaxis_title='ë‚ ì§œ',
    yaxis_title='ì •ê·œí™” ê°€ê²© (ì‹œì‘ì  = 100)',
    hovermode='x unified',
    height=600,
    showlegend=True,
    template='plotly_white'
)

fig_compare.add_hline(y=100, line_dash="dash", line_color="gray",
                       annotation_text="ì‹œì‘ì ", annotation_position="right")

fig_compare.show()

print("âœ“ ë¹„êµ ì°¨íŠ¸ ì™„ì„±")

# ==================== 5ê°œë…„ ì„±ì¥ë¥  ë¶„ì„ (ì£¼ì‹ë§Œ) ====================

print("\n" + "=" * 80)
print("ğŸ“ˆ MAG 7 (ì£¼ì‹) - ìµœê·¼ 5ê°œë…„ ì„±ì¥ë¥  ë¶„ì„")
print("=" * 80)

def get_5year_growth_metrics(ticker):
    """5ê°œë…„ ì„±ì¥ë¥  ë°ì´í„° ìˆ˜ì§‘"""
    try:
        stock = yf.Ticker(ticker)
        financials = stock.financials
        balance_sheet = stock.balance_sheet
        cashflow = stock.cashflow

        if financials.empty:
            return None

        years = financials.columns[:5] if len(financials.columns) >= 5 else financials.columns

        growth_data = {
            'Ticker': ticker,
            'Years': [year.year for year in years],
            'Revenue': [],
            'Operating_Income': [],
            'Net_Income': [],
            'Free_Cash_Flow': []
        }

        if 'Total Revenue' in financials.index:
            growth_data['Revenue'] = financials.loc['Total Revenue', years].tolist()

        if 'Operating Income' in financials.index:
            growth_data['Operating_Income'] = financials.loc['Operating Income', years].tolist()

        if 'Net Income' in financials.index:
            growth_data['Net_Income'] = financials.loc['Net Income', years].tolist()

        if 'Free Cash Flow' in cashflow.index:
            growth_data['Free_Cash_Flow'] = cashflow.loc['Free Cash Flow', years].tolist()

        return growth_data

    except Exception as e:
        return None

def calculate_cagr(start_value, end_value, years):
    """CAGR ê³„ì‚°"""
    if start_value <= 0 or end_value <= 0 or years <= 0:
        return 0
    try:
        cagr = (pow(end_value / start_value, 1 / years) - 1) * 100
        return round(cagr, 2)
    except:
        return 0

print("\n5ê°œë…„ ì¬ë¬´ ë°ì´í„° ìˆ˜ì§‘ ì¤‘ (ì£¼ì‹ë§Œ)...\n")

all_growth_data = []
stock_tickers = [t for t in all_tickers if MAG7_STOCKS[t]['type'] == 'Stock']

for ticker in stock_tickers:
    print(f"ìˆ˜ì§‘ ì¤‘: {ticker}...", end=" ")
    growth_data = get_5year_growth_metrics(ticker)

    if growth_data:
        all_growth_data.append(growth_data)
        print(f"âœ“ ({len(growth_data['Years'])}ê°œë…„)")
    else:
        print("âœ—")

if all_growth_data:
    print(f"\nâœ“ ì´ {len(all_growth_data)}ê°œ ì¢…ëª© ë°ì´í„° ìˆ˜ì§‘ ì™„ë£Œ\n")

    # ë§¤ì¶œ ì„±ì¥ ì°¨íŠ¸
    fig_revenue = go.Figure()

    for data in all_growth_data:
        if data['Revenue']:
            revenue_billions = [val / 1e9 for val in reversed(data['Revenue'])]
            years_reversed = list(reversed(data['Years']))

            fig_revenue.add_trace(go.Scatter(
                x=years_reversed,
                y=revenue_billions,
                mode='lines+markers',
                name=data['Ticker'],
                line=dict(width=3),
                marker=dict(size=10)
            ))

    fig_revenue.update_layout(
        title='ğŸ“Š MAG 7 - 5ê°œë…„ ë§¤ì¶œ ì„±ì¥ ì¶”ì´',
        xaxis_title='ì—°ë„',
        yaxis_title='ë§¤ì¶œ (Billion USD)',
        hovermode='x unified',
        height=600,
        template='plotly_white'
    )

    fig_revenue.show()

    # ìˆœì´ìµ ì„±ì¥ ì°¨íŠ¸
    fig_income = go.Figure()

    for data in all_growth_data:
        if data['Net_Income']:
            income_billions = [val / 1e9 for val in reversed(data['Net_Income'])]
            years_reversed = list(reversed(data['Years']))

            fig_income.add_trace(go.Scatter(
                x=years_reversed,
                y=income_billions,
                mode='lines+markers',
                name=data['Ticker'],
                line=dict(width=3),
                marker=dict(size=10)
            ))

    fig_income.update_layout(
        title='ğŸ’° MAG 7 - 5ê°œë…„ ìˆœì´ìµ ì„±ì¥ ì¶”ì´',
        xaxis_title='ì—°ë„',
        yaxis_title='ìˆœì´ìµ (Billion USD)',
        hovermode='x unified',
        height=600,
        template='plotly_white'
    )

    fig_income.show()

    # CAGR ê³„ì‚°
    print("\n" + "=" * 80)
    print("ğŸ“Š MAG 7 - 5ê°œë…„ ì—°í‰ê·  ì„±ì¥ë¥  (CAGR)")
    print("=" * 80)

    cagr_summary = []

    for data in all_growth_data:
        ticker = data['Ticker']
        years_count = len(data['Years']) - 1

        if years_count <= 0:
            continue

        cagr_data = {'Ticker': ticker}

        if data['Revenue'] and len(data['Revenue']) >= 2:
            cagr_data['ë§¤ì¶œ_CAGR_%'] = calculate_cagr(
                data['Revenue'][-1], data['Revenue'][0], years_count)
        else:
            cagr_data['ë§¤ì¶œ_CAGR_%'] = 'N/A'

        if data['Net_Income'] and len(data['Net_Income']) >= 2:
            start_ni = data['Net_Income'][-1]
            end_ni = data['Net_Income'][0]
            if start_ni > 0 and end_ni > 0:
                cagr_data['ìˆœì´ìµ_CAGR_%'] = calculate_cagr(start_ni, end_ni, years_count)
            else:
                cagr_data['ìˆœì´ìµ_CAGR_%'] = 'N/A'
        else:
            cagr_data['ìˆœì´ìµ_CAGR_%'] = 'N/A'

        cagr_data['ë¶„ì„ê¸°ê°„'] = f"{data['Years'][-1]}-{data['Years'][0]}"
        cagr_summary.append(cagr_data)

    df_cagr = pd.DataFrame(cagr_summary)
    print("\n")
    print(df_cagr.to_string(index=False))

    # CAGR ë¹„êµ ì°¨íŠ¸
    fig_cagr = go.Figure()

    revenue_cagr = []
    income_cagr = []
    tickers_cagr = []

    for _, row in df_cagr.iterrows():
        if row['ë§¤ì¶œ_CAGR_%'] != 'N/A':
            revenue_cagr.append(row['ë§¤ì¶œ_CAGR_%'])
            tickers_cagr.append(row['Ticker'])

    fig_cagr.add_trace(go.Bar(
        name='ë§¤ì¶œ CAGR',
        x=tickers_cagr,
        y=revenue_cagr,
        text=[f"{v:.1f}%" for v in revenue_cagr],
        textposition='auto',
        marker_color='#2196F3'
    ))

    income_cagr = []
    for _, row in df_cagr.iterrows():
        if row['ìˆœì´ìµ_CAGR_%'] != 'N/A':
            income_cagr.append(row['ìˆœì´ìµ_CAGR_%'])

    if len(income_cagr) == len(tickers_cagr):
        fig_cagr.add_trace(go.Bar(
            name='ìˆœì´ìµ CAGR',
            x=tickers_cagr,
            y=income_cagr,
            text=[f"{v:.1f}%" for v in income_cagr],
            textposition='auto',
            marker_color='#FF9800'
        ))

    fig_cagr.update_layout(
        title='ğŸ“Š MAG 7 - 5ê°œë…„ ì—°í‰ê·  ì„±ì¥ë¥  (CAGR) ë¹„êµ',
        xaxis_title='ì¢…ëª©',
        yaxis_title='CAGR (%)',
        barmode='group',
        height=600,
        template='plotly_white'
    )

    fig_cagr.show()

# ìµœì¢… ìš”ì•½
print("\n" + "=" * 80)
print("ğŸ“ˆ MAG 7 + BTC íˆ¬ì ì „ëµ ìµœì¢… ìš”ì•½")
print("=" * 80)

print(f"\nğŸ“ ë¶„ì„ ê¸°ì¤€: Anchored VWAP ({quarter_start.year} Q{quarter_num})")
print(f"ğŸ“… Anchor Point: {quarter_start.strftime('%Y-%m-%d')}")
print(f"ğŸŒŸ ë¶„ì„ ëŒ€ìƒ: MAGNIFICENT SEVEN + BITCOIN (8ê°œ ìì‚°)")

above_vwap = df_results[df_results['Is_Above_VWAP'] == True]
print(f"\nâœ… VWAP ìœ„ ìì‚°: {len(above_vwap)}ê°œ")

strong_buy = df_results[df_results['Buy_Signal_Score'] >= 80]
print(f"ğŸ’š ê°•ë ¥ ë§¤ìˆ˜ (80ì â†‘): {len(strong_buy)}ê°œ")
if not strong_buy.empty:
    tickers_list = [f"{'â‚¿' if MAG7_STOCKS[t]['type'] == 'Crypto' else 'ğŸ“ˆ'} {t}"
                    for t in strong_buy['Ticker'].tolist()]
    print(f"   â†’ {', '.join(tickers_list)}")

good_buy = df_results[(df_results['Buy_Signal_Score'] >= 60) &
                      (df_results['Buy_Signal_Score'] < 80)]
print(f"ğŸ’› ëˆŒë¦¼ëª© ëŒ€ê¸° (60-80ì ): {len(good_buy)}ê°œ")
if not good_buy.empty:
    tickers_list = [f"{'â‚¿' if MAG7_STOCKS[t]['type'] == 'Crypto' else 'ğŸ“ˆ'} {t}"
                    for t in good_buy['Ticker'].tolist()]
    print(f"   â†’ {', '.join(tickers_list)}")

watch = df_results[df_results['Buy_Signal_Score'] < 60]
print(f"ğŸ’™ ê´€ì°° í•„ìš” (60ì  ë¯¸ë§Œ): {len(watch)}ê°œ")
if not watch.empty:
    tickers_list = [f"{'â‚¿' if MAG7_STOCKS[t]['type'] == 'Crypto' else 'ğŸ“ˆ'} {t}"
                    for t in watch['Ticker'].tolist()]
    print(f"   â†’ {', '.join(tickers_list)}")

print("\nğŸ† TOP 3 ì¶”ì²œ:")
for idx in range(min(3, len(df_results))):
    row = df_results.iloc[idx]
    icon = "â‚¿" if row['Type'] == 'Crypto' else "ğŸ“ˆ"
    print(f"  {idx+1}. {icon} {row['Ticker']} - {row['Company']}")
    print(f"     ğŸ’° ì‹œì´: ${row['Market_Cap_Trillion']:.2f}T | " +
          f"ì ìˆ˜: {row['Buy_Signal_Score']}/100 | " +
          f"ë¶„ê¸°ìˆ˜ìµ: {row['Quarter_Return_%']:+.2f}%")

# ì£¼ì‹ vs ì•”í˜¸í™”í ë¹„êµ
stocks_df = df_results[df_results['Type'] == 'Stock']
crypto_df = df_results[df_results['Type'] == 'Crypto']

print(f"\nğŸ“Š ìì‚°êµ°ë³„ í‰ê·  ì„±ê³¼:")
print(f"  ğŸ“ˆ ì£¼ì‹ (MAG 7):")
print(f"     í‰ê·  ë¶„ê¸° ìˆ˜ìµë¥ : {stocks_df['Quarter_Return_%'].mean():.2f}%")
print(f"     í‰ê·  ë§¤ìˆ˜ ì ìˆ˜: {stocks_df['Buy_Signal_Score'].mean():.1f}/100")

if not crypto_df.empty:
    print(f"  â‚¿ ì•”í˜¸í™”í (BTC):")
    print(f"     ë¶„ê¸° ìˆ˜ìµë¥ : {crypto_df['Quarter_Return_%'].iloc[0]:.2f}%")
    print(f"     ë§¤ìˆ˜ ì ìˆ˜: {crypto_df['Buy_Signal_Score'].iloc[0]}/100")

print("\nğŸ’¡ íˆ¬ì ê°€ì´ë“œ:")
print("""
1. ğŸ’š ê°•ë ¥ ë§¤ìˆ˜ (80ì  ì´ìƒ):
   - VWAP ìœ„ì—ì„œ ì•ˆì •ì , ì¦‰ì‹œ ë§¤ìˆ˜ ê²€í† 
   - VWAP +5% ì´ìƒì´ë©´ ëˆŒë¦¼ëª© ëŒ€ê¸° ê¶Œì¥

2. ğŸ’› ëˆŒë¦¼ëª© ëŒ€ê¸° (60-80ì ):
   - VWAP ê·¼ì²˜ ì¡°ì • ì‹œ ë§¤ìˆ˜ ê¸°íšŒ
   - ì†ì ˆë¼ì¸: VWAP -2%

3. ğŸ’™ ê´€ì°° í•„ìš” (60ì  ë¯¸ë§Œ):
   - VWAP ëŒíŒŒ í™•ì¸ í›„ ì¬ê²€í† 
   - í€ë”ë©˜í„¸ê³¼ ë³‘í–‰ ë¶„ì„ í•„ìˆ˜

4. â‚¿ ë¹„íŠ¸ì½”ì¸ íŠ¹ë³„ ê³ ë ¤ì‚¬í•­:
   - ë³€ë™ì„±ì´ ì£¼ì‹ ëŒ€ë¹„ ë†’ìŒ
   - í¬íŠ¸í´ë¦¬ì˜¤ì˜ 5-10% ê¶Œì¥
   - ì¥ê¸° ê´€ì  ì ‘ê·¼ (ë””ì§€í„¸ ê³¨ë“œ)
""")

print("=" * 80)
print("âœ¨ MAGNIFICENT SEVEN + BITCOIN ë¶„ì„ ì™„ë£Œ!")
print("=" * 80)
```
